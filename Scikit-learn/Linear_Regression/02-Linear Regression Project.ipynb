* Avg. Session Length: Average session of in-store style advice sessions.

## Training and Testing Data

Now that we've explored the data a bit, we will split the data into training and testing sets.
We will Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column.

```python
customers.columns
```

```python
X = customers[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]

y = customers['Yearly Amount Spent']
```

** We will use model_selection.train_test_split from sklearn to split the data into training and testing sets.We will set test_size=0.3 and random_state=101**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training the Model

Now its time to train our model on our training data!

```python
from sklearn.linear_model import LinearRegression
```

**Create an instance of a LinearRegression() model named lm.**

```python
lm = LinearRegression()
```

** Train/fit lm on the training data.**

```python
lm.fit(X_train, y_train)
```

**Print out the coefficients of the model**

```python
print('Coefficinets: ', +lm.coef_)
```

```python
coef = pd.DataFrame(lm.coef_, X.columns, columns= ['Coefficients'])
coef
```

## Predicting Test Data
Now that we have fit our model, we will evaluate its performance by predicting off the test values!

** We will use lm.predict() to predict off the X_test set of the data.**

```python
predictions= lm.predict(X_test)
```

** And we will create a scatterplot of the real test values versus the predicted values. **

```python
plt.pyplot.scatter(y_test, predictions)
plt.pyplot.xlabel('Real Test Values')
plt.pyplot.ylabel('Predicted Values')
```

## Evaluating the Model

We will evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

We will calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error. 

```python
from sklearn import metrics
```

```python
print('MAE: {}'.format(metrics.mean_absolute_error(y_test, predictions)))
print('MSE: {}'.format(metrics.mean_squared_error(y_test, predictions)))
print('RMSE: {}'.format(np.sqrt(metrics.mean_squared_error(y_test, predictions))))
```

**The explained variance score** explains the dispersion of errors of a given dataset, and the formula is written as follows: Here, and Var(y) is the variance of prediction errors and actual values respectively. **Scores close to 1.0 are highly desired, indicating better squares of standard deviations of errors.**

```python
metrics.explained_variance_score(y_test, predictions)
```

## Residuals

We will explore the residuals to make sure everything was okay with our data. 

**We will plot a histogram of the residuals and make sure it looks normally distributed.**

```python
sns.set_style('whitegrid')
sns.distplot(y_test-predictions, bins = 50)
```
